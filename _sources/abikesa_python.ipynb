If you have any questions about the code or need further customization, feel free to ask!

````python
# pip install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('transplant_data.csv')

# Assuming the data file has columns 'Region' and 'Percentage'
data = data.pivot("Region", "Percentage")
# Create a heatmap
plt.figure(figsize=(10, 8))
heatmap = plt.pcolor(data, cmap='Reds')
plt.colorbar(heatmap)

# Setting labels
plt.xticks(range(len(data.columns)), data.columns)
plt.yticks(range(len(data.index)), data.index)
plt.title('Heatmap of Heart Transplantations by UNOS Region')

# Show the plot
plt.show()
````
